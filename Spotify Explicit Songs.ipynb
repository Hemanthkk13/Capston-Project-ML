```python
import random
import numpy as np
import pandas as pd
import seaborn as sns
import missingno as msno
import matplotlib.pyplot as plt
```

3 . Data Description

Source - https://www.kaggle.com/datasets/nelgiriyewithana/most-streamed-spotify-songs-2024

4 . DATA IMPORT

```python
df = pd.read_csv("Most Streamed Spotify Songs 2024.csv",encoding = 'latin1')
df
```

5 . Data Cleaning

```python
df.describe()
```

```python
df.info()
```

5.1 Missing Values

```python
msno.bar(df)
plt.show()
```

6 . Feature Engineering

```python
def feature_engineering(df):
    df['Release Date'] = pd.to_datetime(df['Release Date'], format='%m/%d/%Y')
    
    df['Year'] = df['Release Date'].dt.year
    df = df.drop(columns=["TIDAL Popularity"])
    
    return df

df = feature_engineering(df)
df.describe()
```

7 . Exploratory Data Analysis(EDA)

7.1  Average Track Score of Each Year

```python
avg_track_score_per_year = df.groupby('Year')['Track Score'].mean()

plt.figure(figsize=(15,10))
avg_track_score_per_year.plot(kind='line',marker='o')
plt.title('Change in Track Score Throught the Years')
plt.xlabel('Year')
plt.ylabel('Average Track Score')
plt.grid(True)
plt.show()
```

7.2 The Best Track Score of Each Year

```python
best_track_score_per_year = df.groupby('Year')['Track Score'].max()

plt.figure(figsize=(15,10))
best_track_score_per_year.plot(kind = 'line',marker='o')
plt.title('Best Track Scores Throughtout the Years')
plt.xlabel('Year')
plt.ylabel('Best Track Score')
plt.grid(True)
plt.show()
```

7.3 The Worst Track Score for Each Year

```python
worst_track_score_per_year = df.groupby('Year')['Track Score'].min()

plt.figure(figsize=(15,10))
worst_track_score_per_year.plot(kind='line',marker='o')
plt.title('Worst Track Scores Throughout the Years')
plt.xlabel('Year')
plt.ylabel('Worst Track Score')
plt.grid(True)
plt.show()
```

7.4 Number of Tracks every Year 

```python
tracks_per_year  =df.groupby('Year')['Track'].count()

fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(14,6))

#Bar Chart
tracks_per_year.plot(kind='bar',ax=axes[0],color='Red')
axes[0].set_title('Number of Tracks every year')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Number of Tracks')

#Line Chart
tracks_per_year.plot(kind='line',ax=axes[1],color='blue',marker='o')
axes[1].set_title('Number of Tracks every year')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Number of Tracks')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

7.5 Correlation Heatmap

```python
numerical_cols = [col for col in df.columns if df[col].dtype not in ['object','category','datatime64[ns]']]
numerical_cols
```

```python
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(14, 7))
sns.heatmap(correlation_matrix, annot=True, linewidths=0.5)
plt.title('Correlation Heatmap',fontsize=20)

plt.show()
```

7.6 Interest In Explicit Music

```python
explicit_track_scores = df.groupby('Explicit Track')['Track Score'].mean()

plt.figure(figsize=(15,10))
explicit_track_scores.plot(kind='bar',color='orange')
plt.title('Track Score depending on Explicitness')
plt.xlabel('Explicit')
plt.ylabel('Average Track Score')
plt.show()
```

7.7 Number of Friendly Track and Explicit Track

```python
explicit_track_scores = df.groupby('Explicit Track')['Track'].count()

plt.figure(figsize=(15,10))
explicit_track_scores.plot(kind='bar',color='blue' )
plt.title('Number of Friendly and Explicit Tracks')
plt.xlabel('Explicit')
plt.ylabel('Number of Tracks')
plt.show()
```

7.8 Track Score vs Explicit Track

```python
fig, ax = plt.subplots(figsize=(10, 8))
sns.scatterplot(x='Track Score', y='Explicit Track',data=df,ax=ax)
ax.set_xlabel('Track Score')
ax.set_ylabel('Explicit Track')
```

7.9 Spotify Streams vs Explicit Track

```python
fig, ax = plt.subplots(figsize=(5, 5))
sns.scatterplot(x='Spotify Streams', y='Explicit Track', data=df, ax=ax)
ax.set_xlabel('Spotify Streams')
ax.set_ylabel('Explicit Track')
```

7.10 Spotify Popularity vs Explicit Track

```python
fig, ax = plt.subplots(figsize=(10, 8))
sns.scatterplot(x='Spotify Popularity', y='Explicit Track', data=df, ax=ax)
ax.set_xlabel('Spotify Popularity')
ax.set_ylabel('Explicit Track')
```

7 . 11 Change in Explicitness Throught the Years

```python
explicit_tracks = df[df['Explicit Track'] == 1]
explicit_tracks_per_year = explicit_tracks.groupby('Year')['Track'].count()

plt.figure(figsize=(15,10))
explicit_tracks_per_year.plot(kind='line',marker='o')
plt.title('Number of Explicit Tracks per Year')
plt.xlabel('Year')
plt.ylabel('Number of Tracks')
plt.grid(True)
plt.show()
```
